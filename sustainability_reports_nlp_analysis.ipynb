```python
import fitz
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Extract text from PDFs
def extract_text(pdf_path):
    """
    Extracts raw text from a PDF file using PyMuPDF.
    """
    doc = fitz.open(pdf_path)
    text = ""

    for page in doc:
        text += page.get_text()

    return text

# Load the reports
corpus = [
    extract_text("abbreport.pdf"),
    extract_text("amazonreport.pdf"),
    extract_text("applereport.pdf"),
    extract_text("bpreport.pdf"),
    extract_text("googlereport.pdf"),
    extract_text("microsoftreport.pdf"),
    extract_text("nestlereport.pdf"),
    extract_text("nikereport.pdf"),
    extract_text("shellreport.pdf")
]

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words="english", max_features=1000)
X = vectorizer.fit_transform(corpus)

feature_names = vectorizer.get_feature_names_out()

# Top words per document
print("\n================ TOP WORDS PER DOCUMENT ================\n")

for i, doc in enumerate(X):
    print(f"Document {i+1}:")

    sorted_idx = np.argsort(doc.toarray()).flatten()[::-1][:5]
    top_words = [feature_names[idx] for idx in sorted_idx]

    print(top_words)
    print()

# LDA Topic Modeling
lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(X)

print("\n================ TOPICS (LDA) ================\n")

for idx, topic in enumerate(lda.components_):
    print(f"Topic {idx+1}:")

    top_words_idx = topic.argsort()[-5:]
    top_words = [feature_names[i] for i in top_words_idx]

    print(top_words)
    print()

# Document-topic distribution
doc_topics = lda.transform(X)

print("\n================ DOCUMENT TOPIC MIX ================\n")

for i, dist in enumerate(doc_topics):
    print(f"Document {i+1} topic distribution:")
    print(dist)
    print()

# Final Insights (update after running)
print("\n================ INSIGHTS ================\n")

print("Project 2 Insights:")
print("- Sustainability and ESG terminology appears across all industries")
print("- Energy companies emphasize transition and emissions reduction strategies")
print("- Technology companies focus on renewable energy and efficiency")
print("- Consumer companies highlight supply chains and packaging sustainability")
```
